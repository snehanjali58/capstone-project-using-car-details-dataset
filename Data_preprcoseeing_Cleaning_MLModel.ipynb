```python
#Importing the necessary libararies.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset

df=pd.read_csv('CAR DETAILS.csv')
df.head()
```

```python
# Learning the dataset
print(df.shape)
print(df.columns)
print(df.dtypes)

#Number of rows= 4340 columns=8
```

```python
# Checking for the NULL Values.
df.isnull().sum()
```

```python
#Checking for the duplicates

df.duplicated().sum()
```

```python
#Dropping the duplicates
df.drop_duplicates(inplace=True)

df.duplicated().sum()
```

```python
# Checking the columns with the object dataype & performing the encoding.
df.dtypes
```

```python
# We will perform the encoding onto the fuel, seller_type, transmission, owner

#Fuel Column
print(df['fuel'].value_counts())

from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['fuel']=lb.fit_transform(df['fuel'])

print('After Label Encoding')
print(df['fuel'].value_counts())

#By looking at the code we can say that Disel is encoded as the 1, Petrol is encoded as 4, CNG is encoded as 0, LPG is encoded as the 3,
#Electric is encoded as the 2
```

```python
# Encoding the seller_type
print(df['seller_type'].value_counts())

df['seller_type']=lb.fit_transform(df['seller_type'])

print('After Label Encoding')
print(df['seller_type'].value_counts())

#Individual is encoded as the 1, Dealer encoded as the 0, Trustmark Dealer is encoded as 2.
```

```python
# Encoding the transmission
print(df['transmission'].value_counts())

df['transmission']=lb.fit_transform(df['transmission'])

print('After Label Encoding')
print(df['transmission'].value_counts())

#Manual is encoded as 1, automatic is encoded as 0.
```

```python
# Encoding the owner
print(df['owner'].value_counts())

df['owner']=lb.fit_transform(df['owner'])

print('After Label Encoding')
print(df['owner'].value_counts())

#1st owner is encoded as 0, 2nd owner is encoded as 2, 3rd owner is encoded as 4, 4th & above owner is encoded as 1, Test Drive encoded as 3
```

```python
df.head()
```

```python
#Taking some sample data & storing it into another CSV file.

df_sample=df.sample(20)
df_sample.head()

df_sample.to_csv('samplecar.csv')
```

```python
#Preparing the data set for the Machine Learning.

# Problem statement for the machine learning model is to predict the selling price using the regression algorithm

df_reg=df.drop('name',axis=1) #dropping of the name column
df_reg
```

```python
#spliting the dataset into the dependent & independent variable

x=df_reg.drop('selling_price',axis=1)
y=df_reg['selling_price']
print(x.shape)
print(y.shape)
```

```python
#Fetaure Scaling onto the x (independent data)
col_names=x.columns
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

x_sc=sc.fit_transform(x)
x=pd.DataFrame(x_sc,columns=col_names)
x.head()
```

```python
#Spiliting the data into the training & testing

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=35)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Creating the evalutaion matrix for the regression model
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

def reg_eval(y,ypred):
    print('MAE',mean_absolute_error(y,ypred))
    print('MSE',mean_squared_error(y,ypred))
    print('RMSE',np.sqrt(mean_squared_error(y,ypred)))
    print('R2 score',r2_score(y,ypred))
    print('\n')

def model_score(model):
  print('Training Score',model.score(x_train,y_train))
  print('Testing Score',model.score(x_test,y_test))
```

```python
#Creating the linear regression model & evaluating it

from sklearn.linear_model import LinearRegression
model_lin=LinearRegression()
model_lin.fit(x_train,y_train)

ypred1=model_lin.predict(x_test)

print('Evalutaing the Model\n')
reg_eval(y_test,ypred1)
model_score(model_lin)
```

```python
#Creating the KNN Regression model

from sklearn. neighbors import KNeighborsRegressor
model_knn=KNeighborsRegressor(n_neighbors=11)
model_knn.fit(x_train,y_train)

ypred2=model_knn.predict(x_test)

print('Evalutaing the Model\n')
reg_eval(y_test,ypred2)
model_score(model_knn)
```

```python
#Performing the ridge & lasso regression:-

from sklearn.linear_model import Lasso, Ridge

model_lasso=Lasso(alpha=5)
model_lasso.fit(x_train,y_train)
ypred3=model_lasso.predict(x_test)

print('Evalutaing the Model\n')
reg_eval(y_test,ypred3)
model_score(model_lasso)
```

```python
model_ridge=Ridge(alpha=0.5)
model_ridge.fit(x_train,y_train)
ypred4=model_ridge.predict(x_test)

print('Evalutaing the Model\n')
reg_eval(y_test,ypred4)
model_score(model_ridge)
```

Selecting the best model out of the four:- <br>

1.We are not selecting the linear regssion model beacuse it has less model score than the KNN model.Linear regression model does not have any hyperparameter to tune it.<br>

2.Into the ridge & lasso regression we have tried various values of alpha. But model score does not change & it also has less model score than the KNN Model. <br>

3. Hence, we have selected the KNN regression model as the best fit model for this problem statement.<br>

```python
#Saving the best fit model

import pickle

pickle.dump(model_knn,open('best_model.pkl','wb'))

#Loding the model again

loaded_model=pickle.load(open('best_model.pkl','rb'))

#Aplying the loaded model onto the sample data

df_sample.head()

#Setting x & y

x_sample=df_sample.drop('selling_price',axis=1)
y_sample=df_sample['selling_price']

#Feature sacaling of X
x_sc=sc.fit_transform(x_sample)
x_sample=pd.DataFrame(x_sc,columns=col_names)

#Genrating the predictions using the loaded model

y_pred=loaded_model.predict(x_sample)
print('Below are the predictions\n')
print(y_pred)

print('\nBelow are the actual values\n')
print(y_sample)

#Testing the model

reg_eval(y_sample,y_pred)
print('Testing Score', loaded_model.score(x_sample,y_sample))
```
