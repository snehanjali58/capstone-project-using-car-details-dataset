```python
#Importing the libararies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#importing the data
df=pd.read_csv('CAR DETAILS.csv')
df.head()
```

```python
# Data Preprocessing
print('Count of null values')
print(df.isnull().sum())

print(df.duplicated().sum())

df.drop_duplicates(inplace=True)

print('After removing duplicates')
print(df.duplicated().sum())

#Checking categorical & numerical columns

cat_col=df.select_dtypes(include='object').columns
num_col=df.select_dtypes(exclude='object').columns

print('Below are the categorical columns')
print(cat_col,'\n')

print('Below are the numercial columns')
print(num_col)
```

```python
#Let's perform some univariate Analysis

a1=df['fuel'].value_counts()

plt.bar(a1.index,a1.values,color='cyan',edgecolor='black')
plt.xlabel('type of fuel')
plt.ylabel('count')
plt.title('count of the type of fuel')
plt.xticks(rotation=90)
plt.show()

print('\n\033[1mConclusion :- People mostly uses the disel cars & electrical cars has least preference among the other type of cars.\033[0m')
```

```python
a2=df['seller_type'].value_counts()

plt.barh(a2.index,a2.values,color='blue',edgecolor='black')
plt.xlabel('count')
plt.ylabel('type of seller')
plt.title('count of the seller_type')
plt.xticks(rotation=90)
plt.show()

print('\n\033[1mConclusion :- Trustmark dealer has sold very less cars & people prefer to buy the car individually instead of busying from dealers.\033[0m')
```

```python
a3=df['transmission'].value_counts()

plt.stackplot(a3.index,a3.values,color='lightgreen')
plt.xlabel('type of transmission')
plt.ylabel('count')
plt.title('count of mode of transmission')
plt.show()

print('\n\033[1mConclusion :- Manual Cars are preferred more than automatic.\033[0m')
```

```python
sns.lineplot(x=df['km_driven'],y=df['selling_price'])
plt.title('km_driven vs selling_price')
plt.show()

print('\n\033[1mConclusion :- As the value of km_driven increases the selling price of the car will decrease \033[0m')
```

```python
#Let's perform bivariate analysis.

sns.violinplot(x=df['year'],y=df['selling_price'],color='red')
plt.title('Selling price vs Year')
plt.xticks(rotation=90)
plt.show()

print('\n\033[1mConclusion :- For each & every year there is the outlier value for selling price & from 1997 to 2020 the selling prices has been increased\033[0m')
```

```python
sns.boxplot(y=df['transmission'],x=df['selling_price'],color='red')
plt.title('mode of transmission vs selling price')
plt.show()

print('\n\033[1mConclusion :- Automatic cars has more selling price than manual cars.\033[0m')
```

```python
#EDA using the grupu by

# Find the avearge of the selling_price,km_driven based on year & plot it onto the clustered bar chart

a4=df.groupby(['year'])[['selling_price','km_driven']].mean()

a4.plot(kind='bar')
plt.title('year vs average of selling price & km driven')
plt.xticks(rotation=90)
plt.show()

print('\n\033[1mConclusion :- Average of selling price increases from 1997 to 2020 & avearge of km_driven deacreses from 1997 to 2020\033[0m')
```

```python
#Calculate the avearge of the selling price based onto type of owner & plot it onto the pie chart

a5=df.groupby(['owner'])['selling_price'].mean()

plt.pie(x=a5.values,labels=a5.index,autopct='%.2f%%')
plt.title('Type of owner vs Avearge of selling_price')
plt.show()

print('\n\033[1mConclusion :- Test drive car owner has the highest selling price & then the first owner has the 2nd highest selling price.\033[0m')
```
